```python
 data['FLOODS'].replace(['YES','NO'],[1,0],inplace = True)   #  Level encoding  replacing values in the Floods column (YES, NO) with (1, 0)
```

```python
data.head()
```

```python
x = data.iloc[:,2:14] # Features/columns from jan to dec
x.head() # First n rows of the dataset or 5 by default
```

```python
y = data.iloc[:,-1]  # Target or lables or column of floods
y
```

```python
c = data[['JUN','JUL','AUG','SEP']]
c.hist()
plt.show()
```

```python
# Doing Bar plot shows comparisons among discrete categories for categorical data

mx = data[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']].mean().plot.bar(figsize=(14,6))
plt.xlabel('Month')
plt.ylabel('Monthly Rainfall')
plt.show()
```

```python
# Implementing Data Peprocessing

from sklearn import preprocessing
minmax = preprocessing.MinMaxScaler(feature_range = (0,1))
minmax.fit(x).transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

```python
# Using Decision Tree Model

from sklearn.tree import DecisionTreeClassifier #packages for decision tree
dtc_clf = DecisionTreeClassifier() #model creation
dtc_clf.fit(x_train,y_train) #model training
```

```python
# Printing predicted values

y_pred = dtc_clf.predict(x_test)
y_pred
```

```python
# Actual values vs Predicted values

pd.DataFrame({'actual':y_test,'prediction':y_pred})
```

```python
# Accuracy for Decision Tree Algorithm

from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(y_test,y_pred))
```

```python
# Using Random Forest Model

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 200)
clf.fit(x_train,y_train)
```

```python
# Printing predicted values

y_pred2= clf.predict(x_test)
y_pred2
```

```python
# Accuracy for random forest

from sklearn import metrics
print("Accuracy of Random Forest:",metrics.accuracy_score(y_test,y_pred2))
```
